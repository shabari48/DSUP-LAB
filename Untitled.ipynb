```python
import numpy as np
squr_mat=np.array([[1 ,2 ,3],[3 ,4,5],[3 ,7,8]])
print(squr_mat)
```

```python
dia=squr_mat.diagonal()
print(dia)
```

```python
idem=np.eye(squr_mat.shape[0])
print(idem)
```

```python
orthogonal_matrix = np.array([[0, 1, 0],
                              [1, 0, 0],
                              [0, 0, -1]])
print(orthogonal_matrix)
```

```python
det=np.linalg.det(squr_mat)
print(det)
```

```python
invers=np.linalg.inv(squr_mat)
print(invers)
```

```python
m,v=np.linalg.eig(squr_mat)
print("The eigen value is")
print(m)
print("the eigen vector is")
print(v)
```
